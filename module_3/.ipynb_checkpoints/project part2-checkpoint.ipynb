```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import re
%matplotlib inline
```

```python
df = pd.read_csv('main_task.csv')
df.columns = ['Restaurant_id', 'City', 'Cuisine_Style', 'Ranking', 'Rating',
       'Price_Range', 'Number_of_Reviews', 'Reviews', 'URL_TA', 'ID_TA']
df.head()
```

```python
df.info()
```

```python
df.loc[df['Price_Range'] == '$$$$', ['Price_Range']] = 4
df.loc[df['Price_Range'] == '$$ - $$$', ['Price_Range']] = 2.5
df.loc[df['Price_Range'] == '$', ['Price_Range']] = 1
df['Price_Range'] = df['Price_Range'].fillna(df['Price_Range'].median())
df['Price_Range'].value_counts()
```

```python
df['Cuisine_Style']= df['Cuisine_Style'].str.replace(r'[\[\]\']','')
df['Cuisine_Style'] = df['Cuisine_Style'].fillna('Unspecified')
value_c_style = pd.DataFrame(df['Cuisine_Style'].str.split(', ').tolist()).stack().value_counts()
value_c_style
```

```python
df.head()
```

```python
df['count_cuisine'] = df.Cuisine_Style.apply(lambda x: len(x.split(',')))
df['count_cuisine'].mean()
```

```python
s = df['Reviews'].str.split(r"'\], \['",expand=True)
s.columns = ['comments','dates']
s1 = s['comments'].str.split(r"', '",expand=True)
s1.columns = ['comment1','comment2']
s2 = s['dates'].str.split(r"', '",expand=True)
s2.columns = ['date1','date2']
s3 = pd.concat([s1,s2],axis=1)
s3.comment1 = s3.comment1.str.replace(r'\[\[\], \[\]\]','None')
s3.comment1 = s3.comment1.str.replace(r'[\[\]\']','')
s3.date1 = s3.date1.str.replace(r'[\[\]\']','')
s3.date2 = s3.date2.str.replace(r'[\[\]\']','')
s3
```

```python
df_new = pd.concat([df,s3],axis=1)
```

```python
df_new['Number_of_Reviews'] = df_new['Number_of_Reviews'].fillna(round(df_new['Number_of_Reviews'].mean(),0))
df_new.info()
```

```python
df_new.drop(['Reviews', 'comment1','comment2'], axis='columns', inplace=True)
df_new
```

```python
df_new['date1'] = pd.to_datetime(df_new['date1'])
df_new['date2'] = pd.to_datetime(df_new['date2'])
```

```python
df_new['diff_date']=df_new['date1']-df_new['date2']
```

```python
df_new.diff_date.max()
```

```python
df_new.diff_date = df_new.diff_date.fillna(df_new.diff_date.mean())
```

```python
df_new.diff_date = df_new.diff_date.dt.days
```

```python
df_new.info()
```

```python
df_new.drop(df_new.select_dtypes(['object']), inplace=True, axis=1)
df_new.head()
```

```python
df_new.info()
```

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
X = df_new.drop(['date1','date2','Rating'], axis = 1)
y = df_new['Rating']
```

```python
# Загружаем специальный инструмент для разбивки:
from sklearn.model_selection import train_test_split
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Создаём, обучаем и тестируем модель

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
